```
# filter rows 
#   isin,  startswith,  endswith,  contains,  like
```

```
from pyspark.sql import SparkSession
from pyspark.sql.functions import col,lit
spark=SparkSession.builder.appName("filter Dataframe").getOrCreate()
```

```
df=spark.read.options(header="True",inferSchema="True").csv('/FileStore/tables/StudentData.csv')
# &--and    | --or
df1=df.filter((col("course")=='DB') & (col("marks")>50))
df1.show()
```

```
l=["DB","DSA"]
df1=df.filter(col("course").isin(l))
df1.show()
```

```
df.filter(col("course").startswith("D"))
df.filter(col("course").endswith("D"))
df.filter(col("course").contains("DB")).show()
```

```
df.filter(col("course").like("%DB%")).show()
```
