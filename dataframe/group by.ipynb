```
from pyspark.sql import SparkSession
from pyspark.sql.functions import col,lit
spark=SparkSession.builder.appName("data frame").getOrCreate()
df=spark.read.options(header="True",inferSchema="True").csv('/FileStore/tables/StudentData.csv')
```

```
df.groupBy("gender","age").sum("marks").show()
# count,max,min,avg,mean
```

```
# multiple aggregration on single dataframe
from pyspark.sql.functions import max,min,avg,count,sum
df.groupBy("course").agg(count('*').alias("total"),sum("marks").alias('marks'),max('marks').alias('maximum_marks')).show()
```

```
# FILTERING AFTER GROUP BY
df.filter(col("gender")=='Male').groupby("course","gender").agg(count('*').alias("total")).filter(col('total')>85).show()
```
