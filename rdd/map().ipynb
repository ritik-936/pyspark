```
from pyspark import SparkConf,SparkContext
```

```
conf=SparkConf().setAppName("Read File")
```

```
sc=SparkContext.getOrCreate(conf=conf)
```

```
rdd=sc.textFile("FileStore/tables/Sample-2.txt")
```

```
rdd.collect()
```

```
rdd2=rdd.map(lambda x:x.split())
```

```
rdd2.collect()
```

```
def func1(x):
    return x.split()
    
```

```
rdd3=rdd.map(func1)
```

```
rdd3.collect()
```

```
def add2(x):
    l=x.split()
    l2=[]
    for i in l:
        l2.append(int(i)+2)
    return l2
```

```
rdd4=rdd.map(add2)
rdd4.collect()
```
